```python
#FIXME
df.loc[(df['type'] == 'Movie') & (df['year_added'] > 2018)]
```

```python
love = df['title'].str.contains("love",case =False)
df_love = df.loc[love]
df_love.head()
```

```python
movies = df[df['type'] == "Movie"].copy()
movies['duration_minutes'] = movies['duration'].str.extract(r'(\d+)\s*min').astype(float)
movies
```

```python
df['listed_in'] = df['listed_in'].astype(str).str.split(', ')
exploded = df.explode("listed_in")
exploded['listed_in'].value_counts().head(10)
```

```python
movies['contenido_largo'] = movies['duration_minutes'].mask(
    movies['duration_minutes'] > 120,
    'Contenido largo'
)
movies
```

```python
movies.loc[(movies['duration_minutes'] > 100) & (movies['rating'] == 'R') & (movies['country'] == "United States")]
```

```python
top10 = movies.nlargest(10, 'duration_minutes')
top10_styled = (
    top10[['title', 'duration_minutes']]
    .style
    .background_gradient(subset=['duration_minutes'], cmap='Reds')
    .format({'duration_minutes': "{:.0f} min"})
    .set_caption("Top 10 películas más largas")
)

top10_styled
```
